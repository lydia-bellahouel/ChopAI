```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

import os
import sys
```

```python
nb_input_images=10 #nb of images to create X & Y --> 9(*100 time periods) for X, 1(*100 time periods) for y
```

```python
for folder in os.listdir("../../data_image"):
    print(len(os.listdir(f"../../data_image/{folder}")))
```

```python
#Consider only folders with at least the required nb of input images
folder_list = [folder for folder in os.listdir("../../data_image") if len(os.listdir(f"../../data_image/{folder}")) >= nb_input_images]

X = []
y = []

for index_folder, folder in enumerate(folder_list):
    folder_image_list = [image for image in os.listdir(f"../../data_image/{folder}")]
    
    if len(folder_image_list) == nb_input_images:
        folder_X_array = np.zeros(((nb_input_images - 1) * 100, 106))
        folder_y_array = np.zeros((100, 106))
        for index_image, image in enumerate(folder_image_list):
            image_array = np.transpose(plt.imread(f"../../data_image/{folder}/{image}"))
            if index_image < (nb_input_images - 1):
                folder_X_array[index_image * 100: (index_image + 1) * 100, :] = image_array
            elif index_image == (nb_input_images - 1):
                folder_y_array = image_array
        print(folder_X_array.shape)
        print(folder_y_array.shape)
        X.append(folder_X_array)
        y.append(folder_y_array)
    
    if len(folder_image_list) >= 2*nb_input_images:
        nb_of_X_for_the_folder = len(folder_image_list)//nb_input_images
        for i in range(0,nb_of_X_for_the_folder):
            folder_X_array = np.zeros(((nb_input_images - 1) * 100, 106))
            folder_y_array = np.zeros((100, 106))
            for index_image, image in enumerate(folder_image_list[i*nb_input_images:(i+1)*nb_input_images]):
                image_array = np.transpose(plt.imread(f"../../data_image/{folder}/{image}"))
                if index_image < ((i+1)*nb_input_images - 1):
                    folder_X_array[index_image * 100: (index_image + 1) * 100, :] = image_array
                elif index_image == ((i+1)*nb_input_images - 1):
                    folder_y_array = image_array
            X.append(folder_X_array)
            y.append(folder_y_array)

X = np.array(X)
y = np.array(y)
```
